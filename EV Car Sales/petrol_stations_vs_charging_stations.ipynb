```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gdf_petrol = gpd.read_file('petrol_stations.geojson')
gdf_ev = gpd.read_file('charging_stations.geojson')
```

```python
print(gdf_petrol.head())
print(gdf_petrol.columns)

print(gdf_ev.head())
print(gdf_ev.columns)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
gdf_petrol.plot(ax=axes[0], marker ='o', markersize = 5, alpha = 0.6, c = 'red', label = 'Petrol Stations')

axes[0].set_title('Petrol Stations in the UK (OSM Data)')
axes[0].set_axis_off()
axes[0].legend(loc='upper left')

gdf_ev.plot(ax=axes[1], marker ='o', markersize = 5, alpha = 0.6, c = 'green', label = 'EV Charging Stations')

axes[1].set_title('EV Charging Stations in the UK (OSM Data)')
axes[1].set_axis_off()
axes[1].legend(loc='upper left')
plt.tight_layout()
plt.show()
```
