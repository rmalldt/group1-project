```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('export.geojson')
```

```python
print(gdf.head())
print(gdf.columns)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, marker ='o', markersize = 5, alpha = 0.6, color = 'blue', label = 'Petrol Stations')

ax.set_title('Petrol Stations in the UK (OSM Data)')
ax.set_axis_off()
plt.legend()
plt.show()
```
