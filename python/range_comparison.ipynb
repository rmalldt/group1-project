```python
import matplotlib.pyplot as plt
import numpy as np
from textwrap import wrap
```

```python
models = ['Tesla Model 3 Performance vs BMW M3 Comp', 'Audi e-tron GT vs Audi RS7', 'Porsche Taycan Turbo GT with Weissach Package vs Porsche 911 GT3 RS']
ev_ranges = [730, 488, 555]
petrol_ranges = [701, 531, 580]
```

```python
x = np.arange(len(models))
width = 0.35
```

```python
wrapped = ['\n'.join(wrap(label, 40)) for label in models]
```

```python
fig, ax = plt.subplots(figsize=(12, 10
                                ))
ax.bar(x - width/2, ev_ranges, width, label='EV*', color='green')
ax.bar(x + width/2, petrol_ranges, width, label='Petrol**', color='red')
ax.set_ylabel('Range (km)')
ax.set_title('Range Comparison of EVs and Petrol Cars')
ax.set_xticks(x)
ax.set_xticklabels(wrapped, rotation = 0)
ax.legend()

fig.subplots_adjust(bottom = 0.25)
fig.text(0.5, 0.02, '* EV range is based on WLTP cycle\n** Petrol range is based on EPA cycle', ha='center', fontsize=10, color='black')
plt.xticks()
plt.subplots_adjust(bottom= 0.1)
plt.show()
```
